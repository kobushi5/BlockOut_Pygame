```python
import pygame as pg
from pygame.locals import *
import sys
import numpy as np
    
pg.init()
clock = pg.time.Clock()

screen = pg.display.set_mode((800,600))
screen_rect = screen.get_rect()
pg.display.set_caption('Block Out')

ball_math = 0.5
bar_math = 1
temp = (0,0)

#ネタ（数学）
def collide(top,bottom,left,right,radius,dx,dy,x,y):
    if (left < x < right and y < top and abs(y - top) < radius) or (left < x < right and y > bottom and abs(y - bottom) < radius):#上面反射
        return dx,-dy
    if (top < y < bottom and x < left and abs(x - left) < radius) or (top < y < bottom and x > right and abs(x - right) < radius):
        return -dx,dy
    if (left-x)**2 + (top-y)**2 < radius**2 or (left-x)**2 + (bottom-y)**2 < radius**2 or (right-x)**2 + (top-y)**2 < radius**2 or (right-x)**2 + (bottom-y)**2 < radius**2:
        return -dx,-dy
    else:
        return dx,dy
    
def cornerCollide(top,bottom,left,right,radius,dx,dy,x,y):
    a = 0
    m = 0
    dx2 = 0
    dy2 = 0
    if  x < left and y < top and (left-x)**2 + (top-y)**2 < radius**2:#左上
        #from IPython.core.debugger import Pdb; Pdb().set_trace()
        a = np.sqrt(dx**2+dy**2)
        m = np.sqrt((left-x)**2 + (top-y)**2) / a
        dx2 = (left-x)/m
        dy2 = (top -y)/m
        return -dx2,-dy2

    if x < left and y > bottom and (left-x)**2 + (bottom-y)**2 < radius**2:#左下
        a = np.sqrt(dx**2+dy**2)
        m = np.sqrt((left-x)**2 + (bottom-y)**2) / a
        dx2 = (left-x)/m
        dy2 = (y-bottom)/m
        return -dx2,dy2
    
    if x > right and y < top and (right-x)**2 + (top-y)**2 < radius**2:#右上
        a = np.sqrt(dx**2+dy**2)
        m = np.sqrt((right-x)**2 + (top-y)**2) / a
        dx2 = (x-right)/m
        dy2 = (top -y)/m
        return dx2,-dy2
    
    if x > right and y > bottom and (right-x)**2 + (bottom-y)**2 < radius**2:#右下
        a = np.sqrt(dx**2+dy**2)
        m = np.sqrt((right-x)**2 + (bottom-y)**2) / a
        dx2 = (x-right)/m
        dy2 = (y-bottom)/m
        return dx2,dy2
    else:
        return dx,dy
    
def main():
    screen.fill((100,10,225))
    status = "IDLE"
    
    score = 0
    life = 3
    message = "PRESS SPACE TO START"
    
    font = pg.font.SysFont(None, 50) #フォント
    font_gameover = pg.font.SysFont(None, 100)
    text_menu = font.render(message,True, (255,0,0))
    screen.blit(text_menu, (screen_rect.centerx - 240, screen_rect.centery))
    
    block = {}
    block_rect = {}
    target = {}
    
    #ブロック生成
    xcnt=0
    ycnt=0
    for i in range(50):
    #    if xcnt == 5:
    #        block[i] = pg.Surface((25,200))
    #        block[i].set_colorkey((0,0,0))
    #        pg.draw.rect(block[i],(255,255,255),(0,0,25,200))
    #        block_rect[i] = block[i].get_rect()
    #        block_rect[i].topleft=(100+(xcnt*60),50+(ycnt*60))
    #        target[i] = True
    #    else:
    #        block[i] = pg.Surface((50,25))
    #        block[i].set_colorkey((0,0,0))
    #        pg.draw.rect(block[i],(255,255,255),(0,0,50,25))
    #        block_rect[i] = block[i].get_rect()
    #        block_rect[i].topleft=(100+(xcnt*60),50+(ycnt*60))
    #        target[i] = True
    
        block[i] = pg.Surface((50,25))
        block[i].set_colorkey((0,0,0))
        pg.draw.rect(block[i],(255,255,255),(0,0,50,25))
        block_rect[i] = block[i].get_rect()
        block_rect[i].topleft=(100+(xcnt*60),50+(ycnt*60))
        target[i] = True
            
        if xcnt ==9:
            ycnt += 1
        if xcnt / 9 == 1:
            xcnt = 0
        else:
            xcnt += 1
            
    #ボール生成
    circ = pg.Surface((50,50))
    circ.set_colorkey((0,0,0))
    pg.draw.circle(circ,(255,255,100),(25,25),25)
    circ_rect = circ.get_rect()
    
    circ_radius = 25
    
    #バー生成
    bar = pg.Surface((100,10))
    bar.set_colorkey((0,0,0))
    pg.draw.rect(bar,(255,255,255),(0,0,100,10))
    bar_rect = bar.get_rect()
    bar_rect.topleft = (screen_rect.centerx - 50,screen_rect.bottom - 50)
    
    print("circ_rect"+str(circ_rect),str(screen_rect))
    
    #初速度
    dx = 5
    dy = 5
    barx = 0
    bary = 0
    collide_flag = 0
    block_collide_flag=0
    
    circ_rect.topleft = (screen_rect.centerx -100,screen_rect.bottom -200)
    
    screen.blit(circ,circ_rect)
    screen.blit(bar,bar_rect)
    pg.display.update()
    
    while True:
        pressed_keys = pg.key.get_pressed()
        if pressed_keys[K_SPACE]:
            status = "START"
            message = ""
            text_menu = font.render(message,True, (255,0,0))
        if not(True in target.values()):
            status = "CLEAR"
        
        if life <=0:
            status = "GAMEOVER"
        
        
        screen.fill((100,10,225))
        screen.blit(text_menu, (screen_rect.centerx - 180, screen_rect.centery))#message表示
        if status=="START":
            text_score = font.render("SCORE: "+ str(score),True, (255,0,0))
            text_life = font.render("LIFE: "+ str(life),True, (255,0,0))
            screen.blit(text_score, (screen_rect.left + 15, screen_rect.top + 15))
            screen.blit(text_life, (screen_rect.right - 170, screen_rect.top + 15))
            
            barx = 0
            bary = 0
            #バー移動
            if pressed_keys[K_LEFT] or pressed_keys[K_a]:
                barx = -10
            if pressed_keys[K_RIGHT] or pressed_keys[K_d]:
                barx = 10
            if pressed_keys[K_UP] or pressed_keys[K_w]:
                bary = -10
            if pressed_keys[K_DOWN] or pressed_keys[K_s]:
                bary = 10
            bar_rect.move_ip(barx,bary)
                
            circ_rect.move_ip(dx,dy)
            
    #バーとの衝突
            
            if collide_flag == 0:
                if bar_rect.colliderect(circ_rect):
                    collide_flag = 1
                    temp = cornerCollide(bar_rect.top,bar_rect.bottom,bar_rect.left,bar_rect.right,25,dx,dy,circ_rect.centerx,circ_rect.centery)
                    if temp[0] != dx or temp[1] != dy:
                        print('corner')
                        dx = temp[0]
                        dy = temp[1]
                        
                    else:
                        print('no corner')
                        dx += barx * 0.4
                        if circ_rect.top < bar_rect.top or circ_rect.bottom > bar_rect.bottom:
                            dy = -dy + bary * 0.4
                        collide_flag = 1
                    
            
#ネタ（数学）
#            temp = collide(bar_rect.top,bar_rect.bottom,bar_rect.left,bar_rect.right,25,dx,dy,circ_rect.centerx,circ_rect.centery)
#            dx, dy = temp[0],temp[1]
            
            bar_rect.clamp_ip(screen_rect)
    
    #ボールと壁の衝突
            if circ_rect.left < screen_rect.left:
                dx = -dx
                collide_flag = 0
            if circ_rect.right > screen_rect.right:
                dx = -dx
                collide_flag = 0
            if circ_rect.top < screen_rect.top:
                dy = -dy
                collide_flag = 0
            if circ_rect.bottom > screen_rect.bottom:
                dy = -dy
                collide_flag = 0
                life -= 1
            circ_rect.clamp_ip(screen_rect)
    
    #        if circ_rect.left < screen_rect.left or circ_rect.right > screen_rect.right:
    #            dx = -dx 
    #        if circ_rect.top < screen_rect.top or circ_rect.bottom > screen_rect.bottom:
    #            dy = -dy
    #        circ_rect.clamp_ip(screen_rect)
    
    #ボールとブロックの衝突
            for i in target:
                if target[i]:
                    if block_rect[i].colliderect(circ_rect):
                        collide_flag = 0
                        
                        temp = cornerCollide(block_rect[i].top,block_rect[i].bottom,block_rect[i].left,block_rect[i].right,25,dx,dy,circ_rect.centerx,circ_rect.centery)
                        if temp[0] != dx or temp[1] != dy:
                            print('corner block')
                            dx = temp[0]
                            dy = temp[1]
                        else:
                            print('no corner block')
                            if block_rect[i].top < circ_rect.centery < block_rect[i].bottom:
                                dx = -dx
                            if block_rect[i].left < circ_rect.centerx < block_rect[i].right:
                                dy = -dy
                        
#                        if circ_rect.left <  block_rect[i].left or circ_rect.right > block_rect[i].right:
#                            dx = -dx
#                            collide_flag = 0
#                        if circ_rect.top < block_rect[i].top or circ_rect.bottom > block_rect[i].bottom:
#                            dy = -dy
#                            collide_flag = 0
                        score += 100
                        target[i] = False
                        break
            if  0 < dy <1:
                dy += 1
            if -1 < dy < 0:
                dy -= 1
                    
            for i in target:
                if target[i]:
                    screen.blit(block[i],block_rect[i]) 
        
            screen.blit(circ,circ_rect)
            screen.blit(bar,bar_rect)
            pg.display.update()
        
        if status == "GAMEOVER":
                screen.fill((0,0,0))
                message = "GAME OVER"
                text_gameover1 = font_gameover.render(message,True, (255,0,0))
                text_gameover2 = font.render("PRESS R TO RETRY",True, (255,0,0))
                text_gameover3 = font.render("PRESS Q TO QUIT",True, (255,0,0))
                screen.blit(text_gameover1, (screen_rect.centerx - 200, screen_rect.centery - 200))
                screen.blit(text_gameover2, (screen_rect.centerx - 180, screen_rect.centery))
                screen.blit(text_gameover3, (screen_rect.centerx - 180, screen_rect.centery + 100))
                screen.blit(text_score, (screen_rect.centerx - 100, screen_rect.centery - 100))
                
                pg.display.update()
                if pressed_keys[K_q]:
                    status = "END"
                if pressed_keys[K_r]:
                    main()
                    return
            
        if status == "CLEAR":
            screen.fill((255,255,255))
            message = "GAME OVER"
            text_gameover1 = font_gameover.render("GAME CLEAR!!",True, (255,0,0))
            text_gameover2 = font.render("PRESS R TO REPLAY",True, (255,0,0))
            text_gameover3 = font.render("PRESS Q TO QUIT",True, (255,0,0))
            screen.blit(text_gameover1, (screen_rect.centerx - 180, screen_rect.centery - 200))
            screen.blit(text_gameover2, (screen_rect.centerx - 180, screen_rect.centery))
            screen.blit(text_gameover3, (screen_rect.centerx - 180, screen_rect.centery + 100))
            screen.blit(text_score, (screen_rect.centerx - 100, screen_rect.centery - 100))
                
            pg.display.update()
            if pressed_keys[K_q]:
                status = "END"
            if pressed_keys[K_r]:
                main()
                return
                
        for event in pg.event.get():
            if event.type == QUIT:
                pg.quit()
                sys.exit()
            
            if status == "END":
                pg.quit()
                sys.exit()
                
        clock.tick(60)
        
main()
```
