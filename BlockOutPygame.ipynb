```python
import pygame as pg
from pygame.locals import *
import sys

pg.init()
clock = pg.time.Clock()
status = "IDLE"


screen = pg.display.set_mode((800,600))
screen_rect = screen.get_rect()
pg.display.set_caption('Block Out')

block = {}
block_rect = {}
target = {}

xcnt=0
ycnt=0
for i in range(50):
    if xcnt == 5:
        block[i] = pg.Surface((25,200))
        block[i].set_colorkey((0,0,0))
        pg.draw.rect(block[i],(255,255,255),(0,0,25,200))
        block_rect[i] = block[i].get_rect()
        block_rect[i].topleft=(100+(xcnt*60),50+(ycnt*60))
        target[i] = True
    else:
        block[i] = pg.Surface((50,25))
        block[i].set_colorkey((0,0,0))
        pg.draw.rect(block[i],(255,255,255),(0,0,50,25))
        block_rect[i] = block[i].get_rect()
        block_rect[i].topleft=(100+(xcnt*60),50+(ycnt*60))
        target[i] = True
    
    if xcnt ==9:
        ycnt += 1
    if xcnt / 9 == 1:
        xcnt = 0
    else:
        xcnt += 1

circ = pg.Surface((300,300))
circ.set_colorkey((0,0,0))
pg.draw.circle(circ,(255,255,100),(150,150),150)
circ_rect = circ.get_rect()

bar = pg.Surface((100,10))
bar.set_colorkey((0,0,0))
pg.draw.rect(bar,(255,255,255),(0,0,100,10))
bar_rect = bar.get_rect()
bar_rect.topleft = (screen_rect.centerx - 50,screen_rect.bottom - 50)

print("circ_rect"+str(circ_rect),str(screen_rect))

dx = 5
dy = 5
circ_rect.topleft = (screen_rect.centerx -10,screen_rect.bottom -200)

while True:
    pressed_keys = pg.key.get_pressed()
    if pressed_keys[K_s]:
        status = "START"
        
    if not(True in target.values()):
        status = "END"
    
    
    screen.fill((100,10,225))
    
    if status=="START":
        
        if pressed_keys[K_LEFT]:
            bar_rect.move_ip(-10,0)
        if pressed_keys[K_RIGHT]:
            bar_rect.move_ip(10,0)
        if pressed_keys[K_UP]:
            bar_rect.move_ip(0,-10)
        if pressed_keys[K_DOWN]:
            bar_rect.move_ip(0,10)
    
        circ_rect.move_ip(dx,dy)
    
        if bar_rect.colliderect(circ_rect):
            if circ_rect.left < bar_rect.left or circ_rect.right > bar_rect.right:
                dx = -dx
            if circ_rect.top < bar_rect.top or circ_rect.bottom > bar_rect.bottom:
                dy = -dy
        bar_rect.clamp_ip(screen_rect)

#        if circ_rect.left < screen_rect.left:
#            dx = -dx + 1
#        if circ_rect.right > screen_rect.right:
#            dx = -dx - 1
#        if circ_rect.top < screen_rect.top:
#            dy = -dy + 1
#        if circ_rect.bottom > screen_rect.bottom:
#            dy = -dy - 1
#        circ_rect.clamp_ip(screen_rect)

        if circ_rect.left < screen_rect.left or circ_rect.right > screen_rect.right:
            dx = -dx
        if circ_rect.top < screen_rect.top or circ_rect.bottom > screen_rect.bottom:
            dy = -dy
        circ_rect.clamp_ip(screen_rect)

        for i in target:
            if target[i]:
                if block_rect[i].colliderect(circ_rect):
                    if circ_rect.left <  block_rect[i].left or circ_rect.right > block_rect[i].right:
                        dx = -dx
                    if circ_rect.top < block_rect[i].top or circ_rect.bottom > block_rect[i].bottom:
                        dy = -dy
                    target[i] = False
                    break
                
        for i in target:
            if target[i]:
                screen.blit(block[i],block_rect[i]) 
    
    screen.blit(circ,circ_rect)
    screen.blit(bar,bar_rect)
    pg.display.update()
    
    for event in pg.event.get():
        if event.type == QUIT:
            pg.quit()
            sys.exit()
        
        if status =="END":
            pg.quit()
            sys.exit()
            
    clock.tick(60)
```
